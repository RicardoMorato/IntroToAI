# Binary Classification

Binary classifiers are supervised learning models trained with labeled data: 0s for the negative class and 1s for the positive. The predictions they make are 0s and 1s too.

Also, they divulge a probability for each class. It is useful to factor into the conclusions made.

```python
import pandas as pd

df = pd.read_csv("/content/Data/titanic.csv")
df.head()
```

```python
df.info()
```

```python
# Get only the needed columns
df = df[["Survived", "Age", "Sex", "Pclass"]]

# One-hot encode the non-numerical columns
df = pd.get_dummies(df, columns=["Sex", "Pclass"])

# Remove null values from the dataframe
df.dropna(inplace=True)

df.head()
```

```python
# Split the dataset for training and testing
from sklearn.model_selection import train_test_split

x = df.drop(["Survived"], axis=1)
y = df["Survived"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=0)
```

```python
# Create a logistic regression model
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state=0)
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
# Cross-validating to get a more precise accuracy score
from sklearn.model_selection import cross_val_score

cross_val_score(model, x, y, cv=5).mean()
```

```python
# Get a confusion matrix to understand how the model performed during testing
%matplotlib inline
from sklearn.metrics import ConfusionMatrixDisplay as cmd

cmd.from_estimator(model, x_test, y_test, display_labels=["Perished", "Survived"], cmap="Blues", xticks_rotation="vertical")
```

## Some discussion points

It is interesting to notice that the model is more accurate predicting that passengers won't survive than when predicting that they will.

This is due to the dataset used to train the model contained more examples of passengers who perished than of passengers who survived.

```python
# Calculating the model's precision, recall, sensitivity and specificity

from sklearn.metrics import precision_score, recall_score

y_pred = model.predict(x_test)

precision = precision_score(y_test, y_pred)

recall = recall_score(y_test, y_pred)
sensitivity = recall # Sensitivity is a synonym for recall

specificity = recall_score(y_test, y_pred, pos_label=0)

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"Sensitivity: {sensitivity}")
print(f"Specificity: {specificity}")
```

```python
# Predict if a 30-year-old female in first class is likely to survive
female = pd.DataFrame(
    {
        "Age": [30],
        "Sex_female": [1],
        "Sex_male": [0],
        "Pclass_1": [1],
        "Pclass_2": [0],
        "Pclass_3": [0]
    }
)

model.predict(female)[0]

probability = model.predict_proba(female)[0][1]
print(f"Probability of survival: {probability:.1%}")
```

```python
# Predict if a 12-year-old boy in second class is likely to survive
boy = pd.DataFrame(
    {
        "Age": [12],
        "Sex_female": [0],
        "Sex_male": [1],
        "Pclass_1": [0],
        "Pclass_2": [1],
        "Pclass_3": [0]
    }
)

model.predict(boy)[0]

probability = model.predict_proba(boy)[0][1]
print(f"Probability of survival: {probability:.1%}")
```

## Credit Card frauds prediction

```python
import pandas as pd

df = pd.read_csv("/content/Data/creditcard.csv")
df.head()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split

x = df.drop(["Time", "Class"], axis=1)
y = df["Class"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression(random_state=0, max_iter=5000)
lr_model.fit(x_train, y_train)
```

```python
%matplotlib inline
from sklearn.metrics import ConfusionMatrixDisplay as cmd

labels = ["Legitimate", "Fraudulent"]
cmd.from_estimator(lr_model, x_test, y_test, display_labels=labels, cmap="Blues", xticks_rotation="vertical")
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(random_state=0)
rf_model.fit(x_train, y_train)

cmd.from_estimator(rf_model, x_test, y_test, display_labels=labels, cmap="Blues", xticks_rotation="vertical")
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gbm_model = GradientBoostingClassifier(random_state=0)
gbm_model.fit(x_train, y_train)

cmd.from_estimator(gbm_model, x_test, y_test, display_labels=labels, cmap="Blues", xticks_rotation="vertical")
```

```python
from sklearn.metrics import precision_score, recall_score

y_pred = rf_model.predict(x_test)

precision = precision_score(y_test, y_pred)

recall = recall_score(y_test, y_pred)
sensitivity = recall

specificity = recall_score(y_test, y_pred, pos_label=0)

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"Sensitivity: {sensitivity}")
print(f"Specificity: {specificity}")
```
